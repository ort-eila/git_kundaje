```python
# we are looking at the following columns: "encode_id" and "tissue name"
for f in files_to_update:
  print(f)
  read_file_to_update_df = read_tsv_file(f)
  print("read_file_to_update_df before adding header")
  print(read_file_to_update_df.head(2))
  # check if the column name available 
  if "TissueName" in read_file_to_update_df.columns:
    print("File {} was already updated. Moving to the next file".format(f))
    continue # make sure that we are not repeating files
  position_to_insert = read_file_to_update_df.columns.get_loc("ManualAnnotationLabel") + 1
  read_file_to_update_df = add_empty_column_at_position(read_file_to_update_df, "TissueName", position_to_insert)
#   read_file_to_update_df = add_empty_column(read_file_to_update_df, "TissueName")
  print("read_file_to_update_df after adding header")
  print(read_file_to_update_df.head(2))
  for index, row in read_file_to_update_df.iterrows():
    print("Index:", index)
    print("Row data:", row)
    print("CellClusterID is:",row['CellClusterID'])
    encode_id = row['CellClusterID'].split("_")[0]
    print("encode_id: ",encode_id)
    ManualAnnotationLabel = row['ManualAnnotationLabel']
    print("ManualAnnotationLabel: ",ManualAnnotationLabel)
#     # read the relevant row from ENCODE
    matching_row = encode_annotation_df.loc[encode_annotation_df['encode_id'] == encode_id]
    tissue_name = matching_row.iloc[0]['tissue name']
    print("Tissue Name:", tissue_name)
    read_file_to_update_df.at[index, 'TissueName'] = tissue_name
    print("read_file_to_update_df.at[index, 'TissueName'] is: ",read_file_to_update_df.at[index, 'TissueName'])
    
    
    output_file_path = f  # Replace this with the actual file path you want to save to
    read_file_to_update_df.to_csv(output_file_path, sep='\t', index=False)
    print("File {} was updated".format(f))
    print("-------------------")

    
```
